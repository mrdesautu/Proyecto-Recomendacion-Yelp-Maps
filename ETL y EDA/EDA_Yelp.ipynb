## LIBRERIAS A UTILIZAR Y LECTURA

```python
import pandas as pd
import geopandas as gpd
import fiona
import folium
```

```python
df = pd.read_pickle('business.pkl')
df.info()
#SE IDENTIFICA LA NECESIDAD DE ELIMINAR LAS COLUMNAS DUPLICADAS, CON VALORES NULOS PREDOMINANTES. 
```

```python
#AVERIGUAR CUALES SON LOS VALORES DENTRO DE CADA COLUMNA.
# LA SIGUIENTE FUNCION NO SE PUEDE REALIZAR SI NO SE ELIMINAN LAS COLUMNAS REPETIDAS.
# df["state"].unique()
df = df.iloc[:, :14] #SE ELIMINAN LAS COLUMNAS. 
df["state"].unique() #VEMOS LOS CODIGOS CORRESPONDIENTES A CADA ESTADO. 
```

```python
#SE FILTRA POR ETIQUETA DE ESTADO (ESTE METODO CUENTA CON MUCHOS ERRORES EN LOS DATOS?)
df_restaurants = df[df['state'] == 'CA']
#SE PUNTUAN LOS VALORES DENTRO DE LA COLUMNA CATEGORIES
df_restaurants["categories"].unique()
```

```python
# MAPA CON LA UBICACION DE LOS DATOS
# Crear un mapa centrado en una ubicación específica
mapa = folium.Map(location=[40.4168, -3.7038], zoom_start=12)

# Añadir marcadores para cada punto en tus datos
for index, row in df_restaurants.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']]).add_to(mapa)

# Mostrar el mapa
mapa
```

## LECTURA ARCHIVO TIPS
SE REALIZO MEDIANTE UN FOR, POR EL TAMAÑO DEL ARCHIVO. 

```python

# Especifica el archivo JSON y el número de líneas a leer en cada iteración
json_file = 'tip.json'
chunksize = 1000

# Inicializa un objeto vacío de pandas para almacenar los datos
df2 = pd.DataFrame()

# Lee el archivo JSON en partes y agrega los datos al DataFrame
for chunk in pd.read_json(json_file, lines=True, chunksize=chunksize):
    df2 = pd.concat([df2, chunk], ignore_index=True)
```
